```python
class members_application:
    
    def __init__(self, filename, filtre):
        data=pd.read_json(filename)
        d=data[filtre].values
        self.ids=self.get_ids(d)
        self.time=self.get_timestamp(d)
        self.item_id=self.get_item_id(d)
        self.user_id=self.get_user_id(d)
        self.title=self.get_title(d)
        self.create_df()
    
    def get_ids(self, x):
        res=[]
        for i in range(len(x)):
            res.append(int(x[i]["id"]))
        return res
    
    def get_timestamp(self, x):
        res=[]
        for i in range(len(x)):
            res.append(x[i]["timestamp"])
        return res
    
    def get_item_id(self, x):
        res=[]
        for i in range(len(x)):
            res.append(int(x[i]["item_id"]))
        return res
    
    def get_user_id(self, x):
        res=[]
        for i in range(len(x)):
            res.append(int(x[i]["user_id"]))
        return res
    
    def get_title(self, x):
        res=[]
        for i in range(len(x)):
            res.append(x[i]["title"])
        return res
    
    def create_df(self):
        dd={}
        dd["id"]=self.ids
        dd["Timestamp"]=self.time
        dd["job_id"]=self.item_id
        dd["user_id"]=self.user_id
        dd["title"]=self.title
        self.df=pd.DataFrame(dd)
```

```python
def get_em_sectors(df_sector, df_em):
    employers=list(df_em["employer_id"].values)
    employers_sectors=list(df_em["employer_sectors_list"].values)
    em_sectors=[]
    for i in range(len(employers)):
        sectors=employers_sectors[i]
        a=""
        for sector in sectors:
            try:
                sector_title=df_sector[df_sector["jobsector_id"]==sector]["jobsector_title"].values[0]
                a+=sector_title+" "
            except:
                pass
        if len(a)<2:
            a="unknown"
        em_sectors.append(a)
    d={}
    d["employer_id"]=employers
    d["employer_sector_title"]=em_sectors
    res=pd.DataFrame(d)
    return res
```
